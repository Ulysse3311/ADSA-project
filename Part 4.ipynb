```python
#or sum(graph[start])-sum([graph[start][u] for u in used])==0
def brutForce1(graph,start,used):
    used.append(start)
    if len(used)==14 :
        return [start]
    else :
        for i in range(14) :
            if i not in used and graph[start][i]:
                return [start]+[brutForce1(graph,i,used)]
```

```python
def brutForce2(graph,curent,usable):
    if usable is None:
        return curent
    else :
        for i in range(14):
            if i in usable and graph[curent][i]:
                u=usable
                return [curent]+[brutForce2(graph,i,u.remove(i))]
```

```python
def brutForce(graph,start,used):
    if len(used)==14 or sum(graph[start])-sum([graph[start][u] for u in used])==0:
        if len(used)==14:
            print(used)
    else :
        for i in range(14) :
            if i not in used and graph[start][i]:
                u=[i for i in used]
                u.append(i)
                brutForce(graph,i,u)
for k in range(14):
    brutForce(graph,k,[k])
```
